```python
data_frame_polynomial = pd.read_csv("Student_Performance.csv")

#polynomial regression part
data_frame_polynomial = data_frame_polynomial.dropna()
data_frame_polynomial = data_frame_polynomial.drop_duplicates()

# converting non numeric data into numeric data
data_frame_polynomial['Extracurricular Activities'] = data_frame_polynomial['Extracurricular Activities'].apply(lambda x:1 if x =='Yes' else 0)


#defining feats and labels
x=data_frame_polynomial[['Hours Studied','Previous Scores','Extracurricular Activities','Sleep Hours','Sample Question Papers Practiced']]
y=data_frame_polynomial['Performance Index']

#setting up the model with the scaling , poly of second degree 
poly_model = make_pipeline(StandardScaler(), PolynomialFeatures(degree=2), LinearRegression())


#corss validation 
kfold = KFold(5,shuffle  = True, random_state=39)
scores = cross_val_score(poly_model, x, y, cv=kfold, scoring="neg_mean_squared_error")

#leave = LeaveOneOut()
#skfold = StratifiedKFold(5,shuffle = True,random_state = 39) 


#evaluating 
print("Score of each fold are  :",scores)
print("Average of scores is  :",scores.mean())
poly_mse = -scores
print("MSE for each fold is :",poly_mse)
print("Average MSE is :",poly_mse.mean())

"""
y_baseline_pred = np.full_like(y, y.mean())
baseline_mse = mean_squared_error(y, y_baseline_pred)
print("Baseline MSE:", baseline_mse)
"""
```
